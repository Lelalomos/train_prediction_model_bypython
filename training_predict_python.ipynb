```python
title_mean_age=[]
title_mean_age.append(list(set(data1.Name))) #à¸”à¸¶à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸¡à¸²à¹€à¸›à¹‡à¸™ set à¹à¸¥à¹‰à¸§à¹à¸›à¸¥à¸‡à¹€à¸›à¹‡à¸™ list à¹à¸¥à¹‰à¸§à¸™à¸³à¹„à¸› à¹à¸›à¸¥à¸‡à¹„à¸›à¹€à¸›à¹‡à¸™ list  
title_mean_age
```

```python
title_mean_age.append(data1.groupby('Name').Age.mean()) # à¸ˆà¸°à¹à¸šà¹ˆà¸‡à¹€à¸›à¹‡à¸™ 2 array à¹‚à¸”à¸¢à¸ˆà¸°à¸¡à¸µà¸•à¸±à¸§à¸—à¸µà¹ˆà¸¡à¸µà¹à¸•à¹ˆà¸Šà¸·à¹ˆà¸­ à¹à¸¥à¸°à¸­à¸µà¸à¸•à¸±à¸§à¸ˆà¸°à¹€à¸›à¹‡à¸™à¸¡à¸µà¸—à¸±à¹‰à¸‡à¸Šà¸·à¹ˆà¸­à¹à¸¥à¸°à¸­à¸²à¸¢à¸¸
title_mean_age[0][4]
```

```python
n_NameAndAge = len(title_mean_age[1])
n_NameAndAge
```

```python
#à¸‡à¸‡à¹†
import numpy as np
n_traning= df.shape[0]   #number of rows
n_NameAndAge = len(title_mean_age[0])
for i in range(0, n_traning):
    if np.isnan(data1.Age[i])==True:
        for j in range(0, n_NameAndAge):
            if data1.Name[i] == title_mean_age[0][j]:
                data1.Age[i] = title_mean_age[1][j]
```

```python
for i in range(0, data1.shape[0]): #à¸à¸²à¸£à¹à¸›à¸¥à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹ƒà¸«à¹‰à¸ªà¸±à¹‰à¸™à¸¥à¸‡à¹€à¸žà¸·à¹ˆà¸­à¸¥à¸”à¸„à¸§à¸²à¸¡à¸‹à¸±à¸šà¸‹à¹‰à¸­à¸™à¸‚à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥  à¸–à¹‰à¸²à¸¡à¸²à¸à¸à¸§à¹ˆà¸² 18 à¹ƒà¸«à¹‰à¹€à¸›à¹‡à¸™ 0 à¸™à¹‰à¸­à¸¢à¸à¸§à¹ˆà¸²à¹€à¸›à¹‡à¸™ 1
    if data1.Age[i] > 18: 
        data1.Age[i]= 0
    else:
        data1.Age[i]= 1
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(penalty='l2',random_state = 0)

#Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10) #à¸à¸³à¸«à¸™à¸”à¸§à¹ˆà¸² X à¸„à¸·à¸­à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ , Y à¸„à¸·à¸­ à¸„à¸³à¸•à¸­à¸š 
print("Logistic Regression:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 9, metric = 'minkowski', p = 2)

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("K-NN:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 2)

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("SVM:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting Naive Bayes to the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("Naive Bayes:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting Random Forest Classification to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 2)

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("Random Forest:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std())
```
