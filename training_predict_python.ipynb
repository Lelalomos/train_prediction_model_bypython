```python
from sklearn.metrics import accuracy_score as AS
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# df = pd.read_csv('./datasets/student_training.csv')
df = pd.read_csv('train.csv')
# labes = df['Major']
df
```

```python
data1 = df.drop(['PassengerId','Cabin','Ticket','Fare', 'Parch', 'SibSp'], axis=1)
y= data1.Survived
y
```

```python
data1
```

```python
from sklearn.preprocessing import LabelEncoder
lableEncode = LabelEncoder() # เอาไว้สำหรับแปลงตัวหนังสือเป็นตัวเลข
```

```python
data1.Sex = lableEncode.fit_transform(data1.Sex) #แปลงจากตัวหนังสือของ Sex เป็นตัวเลข 0,1
data1
```

```python
print ('Number of null values in Embarked:', sum(data1.Embarked.isnull())) #หาค่า null ของ Embarked column ใน data1 ซึ่งก็คือ datafame แล้วนำมานับว่ามีกี่ตัว
```

```python
row_index = data1.Embarked.isnull() #กำหนดให้ตัวแปร row_index เก็บตัวที่ค่าเป็น null
data1.loc[row_index,'Embarked']='S' #เพิ่ม data1 ในช่องที่ว่างเป็น S
data1
```

```python
Embarked  = pd.get_dummies(data1.Embarked , prefix='Embarked'  ) # เป็นการแบ่ง Embarked เป็นแต่ละส่วน
Embarked
```

```python
data1 = data1.drop(['Embarked'], axis=1)
data1
```

```python
data1= pd.concat([data1, Embarked], axis=1)  
data1
```

```python
data1 = data1.drop(['Embarked_S'], axis=1)
```

```python
data1.head()
```

```python
n_row_traning= data1.shape[0] #shape[0] เป็นการบอกจำนวน rows 
n_column_traning = data1.shape[1] #shape[1] เป็นการบอกจำนวน column 
print("จำนวน row : {} จำนวน column : {} ในข้อมูล".format(n_row_traning,n_column_traning))
```

```python
title_mean_age=[]
title_mean_age.append(list(set(data1.Name))) #ดึงข้อมูลมาเป็น set แล้วแปลงเป็น list แล้วนำไป แปลงไปเป็น list  
title_mean_age
```

```python
title_mean_age.append(data1.groupby('Name').Age.mean()) # จะแบ่งเป็น 2 array โดยจะมีตัวที่มีแต่ชื่อ และอีกตัวจะเป็นมีทั้งชื่อและอายุ
title_mean_age[0][4]
```

```python
n_NameAndAge = len(title_mean_age[1])
n_NameAndAge
```

```python
#งงๆ
import numpy as np
n_traning= df.shape[0]   #number of rows
n_NameAndAge = len(title_mean_age[0])
for i in range(0, n_traning):
    if np.isnan(data1.Age[i])==True:
        for j in range(0, n_NameAndAge):
            if data1.Name[i] == title_mean_age[0][j]:
                data1.Age[i] = title_mean_age[1][j]
```

```python
print("จำนวนอายุที่หายไป : {}".format(sum(data1.Age.isnull())))
```

```python
#data1.fillna(0,inplace=true) #ใส่ข้อมูลในช่องที่ว่าง
mean_age = data1.Age.mean() #หาค่าเฉลี่ย Age 
data1.fillna(mean_age,inplace=True) #ใส่ค่าเฉลี่ยที่หาได้ไปใส่ลงช่องที่ว่าง
```

```python
data1=data1.drop(['Name'], axis=1)
```

```python
data1 = data1.drop(['Survived'],axis=1)
data1
```

```python
print("จำนวนอายุที่หายไป : {}".format(sum(data1['Age'].isnull())))
```

```python
for i in range(0, data1.shape[0]): #การแปลงข้อมูลให้สั้นลงเพื่อลดความซับซ้อนของข้อมูล  ถ้ามากกว่า 18 ให้เป็น 0 น้อยกว่าเป็น 1
    if data1.Age[i] > 18: 
        data1.Age[i]= 0
    else:
        data1.Age[i]= 1
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(penalty='l2',random_state = 0)

#Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10) #กำหนดว่า X คือข้อมูล , Y คือ คำตอบ 
print("Logistic Regression:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 9, metric = 'minkowski', p = 2)

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("K-NN:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 2)

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("SVM:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting Naive Bayes to the Training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("Naive Bayes:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std(),"\n")
```

```python
# Fitting Random Forest Classification to the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 2)

# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X=data1 , y=y , cv = 10)
print("Random Forest:\n Accuracy:", accuracies.mean(), "+/-", accuracies.std())
```
